## Importing the necesssary libraries and loading the dataset:
---

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
import warnings
warnings.filterwarnings('ignore')

#   Initialize the Spark session
spark = SparkSession.builder \
    .master("local") \
    .appName("Exercise3") \
    .getOrCreate()

df = spark.read.format('org.apache.spark.sql.json') \
        .load("icc_results.json")
```

## Fetching the description column from the JSON filr :
---

```python
match_s = df.select('scheduleentry.description')
match_s.show()
```

## Splitting the string in 'description' column above to get the mode of match :
---

```python
mode = match_s.withColumn('mode', split(match_s['description'], ' ').getItem(1))
mode= mode.drop('description')
mode.show(200)
```

## Loading the matchId column to fetch out the year :
---

```python
match_id = df.select('scheduleEntry.matchId.name')
match_id.show()
```

## Splitting the string in the 'match_id' column above tp fetch out the year the match was played :
---

```python
year = match_id.withColumn('Year', split(match_id['name'], '-').getItem(2))
year= year.drop('name')

year.show(200)
```

## Merging both the columns obtained to get a dataframe of the mode of match and the year it was played in and also seperating the ODI and Test matches:
---

```python
mode = mode.withColumn("id", monotonically_increasing_id())
year = year.withColumn("id", monotonically_increasing_id())

resultdf = mode.join(year, "id", "outer").drop("id")
resultdf=resultdf.dropna()
odidf= resultdf.filter(resultdf.mode == 'ODI')
testdf=resultdf.filter(resultdf.mode == 'Test')

odidf.show()
testdf.show()
```

## Total number of ODIs played each year :
---

```python
odicount= odidf.groupBy("Year").count()
odicount=odicount.withColumnRenamed("count","Number of ODIs")

odicount.show()
```

## Total number of Test matches played each year:
---

```python
testcount= testdf.groupBy("Year").count()
testcount=testcount.withColumnRenamed("count","Number of Tests")

testcount.show()
```

## Performing an outer join to evalute the number of ODIs and Tests played each year:
--

```python
finaldf = odicount.join(testcount,"Year",'outer')
```
